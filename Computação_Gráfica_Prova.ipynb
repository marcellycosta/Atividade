```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Criação da figura e do eixo
fig, ax = plt.subplots()

# Desenha o círculo de fundo branco
circle = plt.Circle((0.5, 0.5), 0.5, color='white')
ax.add_artist(circle)

# Desenha a faixa verde
ax.add_patch(patches.Rectangle((0.0, 0), 0.333, 1, color='green'))

# Desenha a faixa branca
ax.add_patch(patches.Rectangle((0.333, 0), 0.334, 1, color='white'))

# Desenha a faixa vermelha
ax.add_patch(patches.Rectangle((0.667, 0), 0.333, 1, color='red'))

#Linha da borda
border_circle = plt.Circle((0.5,0.5), 0.5, color='black', fill=False, linewidth=5)
ax.add_patch(border_circle)

# Ajuste dos eixos para não aparecer margens
ax.set_aspect(1)
plt.xlim(0, 1)
plt.ylim(0, 1)

# Remove os eixos
ax.axis('off')

# Adiciona uma máscara circular para cortar as extremidades
clip_circle = patches.Circle((0.5, 0.5), 0.5, transform=ax.transData)
for patch in ax.patches:
    patch.set_clip_path(clip_circle)

# Mostra a bandeira
plt.show()


```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
fig, ax = plt.subplots()

# Círculos cinzas
circle = patches.Circle((0.35, 0.6), 0.06, edgecolor='black', facecolor='gray')
ax.add_patch(circle)
circle = patches.Circle((0.65, 0.6), 0.06, edgecolor='black', facecolor='gray')
ax.add_patch(circle)

# Triangulo rosa
triangle = patches.Polygon([[0.4, 0.4], [0.60, 0.4], [0.5, 0.55]], closed=True,
edgecolor='black', facecolor='pink')
ax.add_patch(triangle)

#Circulo roxo
circle = patches.Circle((0.50, 0.25), 0.09, edgecolor='black', facecolor='purple')
ax.add_patch(circle)

ax.set_aspect('equal')
plt.xlim(0, 1)
plt.ylim(0, 1)
ax.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os valores de x
x = np.linspace(-5, 5, 400)

# Definir as funções y = 2 * sen(x) e y = 2 * cos(x)
y1 = 2 * np.sin(x)
y2 = 2 * np.cos(x)

# Criar a figura e o eixo do gráfico
plt.figure()

# Plotar a função y = 2 * sen(x)
plt.plot(x, y1, label=r'$y = 2 \sin(x)$', color='orange', linestyle='-')

# Plotar a função y = 2 * cos(x)
plt.plot(x, y2, label=r'$y = 2 \cos(x)$', color='blue', linestyle='-')

# Preencher a área entre as funções no intervalo de x entre -2 e 0
plt.fill_between(x, y1, y2, where=(x >= -2) & (x <= 0), color='gray', alpha=0.5)

# Adicionar o título e os rótulos dos eixos
plt.title('Gráfico de y = 2 sen(x) e y = 2 cos(x)')
plt.xlabel('x')
plt.ylabel('y')

# Limitar os valores dos eixos x e y
plt.xlim([-5, 5])
plt.ylim([-2.5, 2.5])

# Adicionar a legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

a = 5
b = 10
x = np.linspace(-a, a, 100)
y = np.linspace(-b, b, 100)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2)
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()

#O código cria um gráfico 3D da função Z= x**2 + y**2 que forma uma superfície em forma de parábola - Não possui erro
```
