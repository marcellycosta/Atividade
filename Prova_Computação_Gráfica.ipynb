```
import matplotlib.pyplot as plt

def desenhar_retangulo(x, y, largura, altura, cor):
    plt.fill([x, x + largura, x + largura, x], [y, y, y + altura, y + altura], color=cor)

def desenhar_lozango(x, y, largura, altura, cor):
    plt.fill([x, x + largura/2, x + largura, x + largura/2],
             [y + altura/2, y, y + altura/2, y + altura], color=cor)

def desenhar_x(x, y, largura, altura, cor, espessura):
    plt.plot([x, x + largura], [y, y + altura], color=cor, linewidth=espessura)
    plt.plot([x + largura, x], [y, y + altura], color=cor, linewidth=espessura)

def desenhar_tabuleiro_invertido(x, y, largura, altura, cor1, cor2, n):
    # Define o tamanho de cada quadrado do tabuleiro
    tamanho = largura / n
    for i in range(n):
        for j in range(n):
            # Alterna entre cor1 e cor2, começando com cor1 (verde)
            cor = cor1 if (i + j) % 2 == 0 else cor2
            desenhar_retangulo(x + i * tamanho, y + j * tamanho, tamanho, tamanho, cor)

# Função para desenhar uma borda ao redor da figura
def desenhar_borda(x, y, largura, altura, cor):
    plt.plot([x, x + largura, x + largura, x, x],
             [y, y, y + altura, y + altura, y], color=cor, linewidth=2)

# Cria uma figura com 3 subplots na horizontal
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Figura 1: Quadrado com losango vermelho
plt.sca(axs[0])
desenhar_retangulo(0, 0, 10, 10, 'white')  # Fundo branco
desenhar_lozango(0, 0, 10, 10, 'red')  # Losango vermelho
desenhar_borda(0, 0, 10, 10, 'black')  # Borda preta
plt.axis('equal')
plt.axis('off')

# Figura 2: Quadrado com "X" azul de linha mais grossa
plt.sca(axs[1])
desenhar_retangulo(0, 0, 10, 10, 'blue')  # Fundo Azul
desenhar_x(0, 0, 10, 10, 'white', espessura=30)  # X azul com linha grossa
desenhar_borda(0, 0, 10, 10, 'black')  # Borda preta
plt.axis('equal')
plt.axis('off')

# Figura 3: Tabuleiro verde e branco, começando com verde
plt.sca(axs[2])
desenhar_tabuleiro_invertido(0, 0, 10, 10, 'white', 'green', 4)  # Tabuleiro invertido
desenhar_borda(0, 0, 10, 10, 'black')  # Borda preta
plt.axis('equal')
plt.axis('off')

# Exibe a figura com as três imagens na horizontal, todas com borda
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Função para plotar um polígono conectando seus pontos com linhas
def plot_polygon_with_lines(points):
  x_coords = [point[0] for point in points]
  y_coords = [point[1] for point in points]

# Plota as linhas do polígono e conecta o último ponto ao primeiro
  plt.plot(x_coords + [x_coords[0]], y_coords + [y_coords[0]], marker='o')
  plt.xlabel("Eixo X")
  plt.ylabel("Eixo Y")
  plt.title("Polígono")
  plt.grid(True)#adicionando grade
  plt.show()

#Função para aplicar uma transformação matricial (escala, rotação, etc.) nos pontos do polígono
def aply_transformation(points, transformation_matrix):
  transformed_points = []
  for point in points:
    if transformation_matrix.shape == (2, 2):
       # Para matriz 2x2, transforma o ponto sem coordenada homogênea
      homogeneous_point = np.array([point[0], point[1]])
      transformed_point = np.dot(transformation_matrix, homogeneous_point)
    elif transformation_matrix.shape == (3, 3):
      # Para matriz 3x3, utiliza coordenadas homogêneas (adiciona 1 ao vetor do ponto)
      homogeneous_point = np.array([point[0], point[1], 1])
      # Divide pelo fator homogêneo (se necessário) para retornar a coordenada 2D
      transformed_point = np.dot(transformation_matrix, homogeneous_point)
      homogeneous_point = transformed_point[:2] / transformed_point[2]
    else:
      raise ValueError("A matrix apresentada deve ser 2x2 ou 3x3") #mensagem de erro
    transformed_points.append((transformed_point[0], transformed_point[1]))
  return transformed_points #Retorna os pontos transformados

# Define os vértices do pentágono como uma lista de tuplas (coordenadas x, y)
vertices_pentagono = [
    (-0.6, -0.75),
    (0.0, 0.30),
    (0.65, 0.50),
    (0.0, 1.0),
    (0.2, 0.10),
    (0.6, -0.70)


]

# Plota o polígono original utilizando a lista de vértices definida anteriormente
plot_polygon_with_lines(vertices_pentagono)

# Define os pontos que serão transformados como os vértices do pentágono
points = vertices_pentagono

# Define o ângulo de rotação em radianos (converte de graus para radianos)
angle = np.radians(-30)

# Cria a matriz de rotação 2x2
# [ cos(θ)  -sin(θ) ]
# [ sin(θ)   cos(θ) ]
rotation_matrix = np.array([
    [np.cos(angle), -np.sin(angle)],
    [np.sin(angle), np.cos(angle)]
])

transformed_point = aply_transformation(points, rotation_matrix)
plot_polygon_with_lines(transformed_point)
```

```
import numpy as np
import matplotlib.pyplot as plt


# Função para calcular uma curva de Bézier
def bezier_curve(points, n_points=100):
    points = np.array(points)
    t = np.linspace(0, 1, n_points)
    n = len(points) - 1
    curve = np.zeros((n_points, 2)) # Inicializa a curva
    for i in range(n + 1):
        # Multiplica cada coordenada dos pontos separadamente
        curve += np.math.comb(n, i) * (1 - t)[:, np.newaxis] ** (n - i) * t[:, np.newaxis] ** i * points[i]
    return curve

# Função para desenhar as curvas
def draw_curve(ax, points, label="", color="black"):
    curve = bezier_curve(points)
    ax.plot(curve[:, 0], curve[:, 1], label=label, color=color)

# Dados
face_outline = [
    [(-2, 4), (-1.5, 4.8), (1.5, 4.8), (2, 4)],   # Testa
    [(-2, 4), (-2.5, 2), (-2.5, -2), (-2, -4)],       # Lado esquerdo
    [(2, 4), (2.5, 2), (2.5, -2), (2, -4)],           # Lado direito
    [(-2, -4), (-1, -4.5), (1, -4.5), (2, -4)]            # Queixo
]

left_eye = [
    [(-2, 2), (-1.5, 2.5), (-0.5, 2.5), (0, 2)],      # Pálpebra superior
    [(-2, 2), (-1.5, 1.8), (-0.5, 1.8), (0, 2)]       # Pálpebra inferior
]

right_eye = [
    [(0, 2), (0.5, 2.5), (1.5, 2.5), (2, 2)],       # Pálpebra superior
    [(0, 2), (0.5, 1.8), (1.5, 1.8), (2, 2)]        # Pálpebra inferior
]

nose = [
    [(0, 2), (-0.3, 1.5), (0.3, 1.5), (0, 1)],   # Ponte do nariz
    [(-0.5, 1), (-0.6, 0.8), (-0.4, 0.8), (-0.5, 1)],      # Narina esquerda
    [(0.5, 1), (0.6, 0.8), (0.4, 0.8), (0.5, 1)]          # Narina direita

]

mouth = [
    [(-1, -1), (-0.5, -0.5), (0.5, -0.5), (1, -1)],   # Lábio superior
    [(-1, -1), (-0.5, -1.5), (0.5, -1.5), (1, -1)]    # Lábio inferior
]

# Configuração do gráfico
fig, ax = plt.subplots(figsize=(6, 8))

# Desenhar contorno do rosto
for segment in face_outline:
    draw_curve(ax, segment, color="brown")

# Desenhar olhos
for segment in left_eye:
    draw_curve(ax, segment, color="blue")
for segment in right_eye:
    draw_curve(ax, segment, color="blue")

# Desenhar nariz
for segment in nose:
    draw_curve(ax, segment, color="black")

# Desenhar boca
for segment in mouth:
    draw_curve(ax, segment, color="red")

# Ajustes de exibição
ax.axis("equal")
ax.axis("off")
plt.title("Rosto desenhado com curvas de Bézier")
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Função para plotar os pontos 3D e suas conexões
def plot_3d_points_with_lines(points, edges=None, labels=None):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")

    x_coords = [point[0] for point in points]
    y_coords = [point[1] for point in points]
    z_coords = [point[2] for point in points]

    # Plotar os pontos
    ax.scatter(x_coords, y_coords, z_coords)

    # Conectar os pontos com linhas
    if edges:
        for edge in edges:
            i, j = edge
            ax.plot(
                [x_coords[i], x_coords[j]],
                [y_coords[i], y_coords[j]],
                [z_coords[i], z_coords[j]],
            )

    if labels:
        for i, label in enumerate(labels):
            ax.text(x_coords[i], y_coords[i], z_coords[i], label)

    ax.set_xlabel("Eixo X")
    ax.set_ylabel("Eixo Y")
    ax.set_zlabel("Eixo Z")
    ax.set_title("Pirâmide 3D Rotacionada")
    plt.show()

# Definir os vértices da pirâmide
vertices = [
    (0, 0, 0),  # Base: Vértice 1
    (1, 0, 0),  # Base: Vértice 2
    (1, 1, 0),  # Base: Vértice 3
    (0, 1, 0),  # Base: Vértice 4
    (0.5, 0.5, 1),  # Topo da pirâmide
]

# Lista de arestas da pirâmide (conexões entre vértices)
edges = [
    (0, 1), (1, 2), (2, 3), (3, 0),  # Arestas da base
    (0, 4), (1, 4), (2, 4), (3, 4),  # Arestas laterais
]

# Função para criar a matriz de rotação
def rotation_matrix_yz(angle_y, angle_z):
    angle_y = np.radians(angle_y)
    angle_z = np.radians(angle_z)

    # Rotação em Y
    rotation_y = np.array([
        [np.cos(angle_y), 0, np.sin(angle_y)],
        [0, 1, 0],
        [-np.sin(angle_y), 0, np.cos(angle_y)],
    ])

    # Rotação em Z
    rotation_z = np.array([
        [np.cos(angle_z), -np.sin(angle_z), 0],
        [np.sin(angle_z), np.cos(angle_z), 0],
        [0, 0, 1],
    ])

    return np.dot(rotation_z, rotation_y)

# Aplicar a matriz de rotação nos vértices
def apply_rotation(vertices, rotation_matrix):
    rotated_vertices = []
    for vertex in vertices:
        rotated_vertex = np.dot(rotation_matrix, np.array(vertex))
        rotated_vertices.append(rotated_vertex)
    return rotated_vertices

# Criar a matriz de rotação para 20° em Y e Z
rotation_matrix = rotation_matrix_yz(20, 20)

# Aplicar a rotação aos vértices
rotated_vertices = apply_rotation(vertices, rotation_matrix)

# Plotar a pirâmide rotacionada
plot_3d_points_with_lines(rotated_vertices, edges=edges)
```
