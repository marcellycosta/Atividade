```python
#Biblioteca
import numpy as np
import matplotlib.pyplot as plt

# Função e derivada
def f(x):
    return np.cos(x)**2 - np.sin(x)

def df(x):
    return -2 * np.cos(x) * np.sin(x) - np.cos(x)

# Intervalo e precisão
a, b = 1.5, 4
tolerancia = 1e-6
inter_max = 100

# Método da Bisseção
def bissecao(f, a, b, tol, max_iter):
    interacao = []
    for i in range(max_iter):
        m = (a + b) / 2
        fm = f(m)
        interacao.append((i, a, b, m, fm))
        if abs(fm) < tol or (b - a) / 2 < tol:
            break
        if np.sign(f(a)) * np.sign(fm) < 0:
            b = m
        else:
            a = m
    return interacao

# Método de Newton-Raphson
def newton_raphson(f, df, x0, tol, max_iter):
    interacao = []
    x = x0
    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)
        interacao.append((i, x, fx, dfx))
        if abs(fx) < tol:
            break
        x -= fx / dfx
    return interacao

# Executar os métodos
inter_bissecao = bissecao(f, a, b, tolerancia, inter_max)
newton_inter = newton_raphson(f, (df), (a+b)/2, tolerancia, inter_max)

# Plotar a função f(x)
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$\cos(x)^2 - \sin(x)$', color='blue')
plt.axhline(0, color='black',linewidth=0.5)
plt.title("Gráfico da função $f(x) = \cos(x)^2 - \sin(x)$")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar o gráfico
plt.show()

inter_bissecao[:10], newton_inter[:10]  # Mostrar primeiras 10 iterações de ambos os métodos
```
